1. A tree is a connected acyclic graph.

Tree is special kind of graph, that is conected (there is a path between any two vertices), acyclic (there are no cycles(A cycle is when you can start at one vertix and, passing along the edges, return to it without repeating the edge.)), and if there are n vertices in a tree, then there are always exactly n - 1 edges. So if we say that something is a tree it has to be a connected acyclic graph.

2. A tree is one component of a forest. (A forest is an acyclic graph.)

A forest is a collection of trees that are not interconnected. If each component of the forest is a tree, and the tree is an acyclic graph, then the forest is also acyclic.

3. A tree is a connected graph with at most V - 1 edges.

Yeap, if there is V edges, graph will become cyclic, which is not tree.

4. A tree is a minimally connected graph; removing any edge disconnects the graph.

In a tree, there are two types of vertices.
The first type is internal vertices — they have two or more edges that lead either to other internal vertices or to leaves.
If we remove one of these edges, we will disconnect either a leaf or another internal vertex from the tree.
The second type is leaves — they have only one edge.
If we remove that edge, the leaf will become disconnected from the tree.

5. A tree is an acyclic graph with at least V - 1 edges.

If there is less than V-1 edges, graph will be disconnected, this kind of graph can't be a tree, so tree must have V-1 edges.

6. A tree is a maximally acyclic graph; adding an edge between any two vertices creates a cycle.

If we don't add new vertex, all new edges lead to cycle.

7. A tree is a graph that contains a unique path between each pair of vertices.

Tree is connected, which means we can find a path between any 2 vertex, and tree is acyclic, which means between any 2 vertex there is only one unique path.
