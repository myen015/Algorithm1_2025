### Problem 1: P, NP, stuff

Ok so for this one, we gotta classify these problems into P, NP, NP-complete, NP-hard. Each line is the same class.

1. find max, linear search, shortest path in unweighted graph, matrix multiplication  
   These are all like basic algos that run in poly time. Finding max is O(n), linear search same, shortest path with BFS is O(V+E), matrix mult is O(n^3) or whatever. So P.

2. sorting of list, Dijkstra on non-negative weights, BFS, DFS, merge sort, quicksort  
   Sorting is O(n log n), Dijkstra with heap is O(E log V), BFS/DFS O(V+E), mergesort and quicksort are O(n log n) avg. All poly time, P.

3. sudoku  
   Sudoku is NP-complete cuz you can reduce it from 3-SAT or coloring. In NP cause verify solution quick, NP-hard cause others reduce to it. So NP, NP-complete, NP-hard.

4. 3 coloring of graph, scheduling with conflicts  
   3-coloring is classic NP-complete. Scheduling with conflicts is basically graph coloring too. Both NP (verify coloring fast), NP-hard.

5. Traveling Salesperson Problem, Hamiltonian Cycle, Clique  
   All NP-complete. TSP from Ham Cycle, Ham Cycle from Vertex Cover or something, Clique also. In NP, NP-hard.

6. Cryptography, factoring large integers  
   Factoring is in NP (check factor quick), but not known if P or complete. Prob intermediate. Crypto based on that. So NP.

7. Halting Problem, busy beaver  
   Halting is undecidable, but NP-hard cause SAT reduces to it. Busy beaver also undecidable and hard. NP-hard.



### Problem 2: 

```
def bayes_calc(prev=0.001, sens=0.99, spec=0.99):
    fp = 1 - spec
    p_pos = (sens * prev) + (fp * (1 - prev))
    post = (sens * prev) / p_pos
    return post

print(f"Prob: {bayes_calc()*100:.1f}%")

# sim
import random

def sim_bayes(n=100000, prev=0.001, sens=0.99, spec=0.99):
    dis = int(n * prev)
    heal = n - dis
    tp = int(dis * sens)
    fp = int(heal * (1 - spec))
    tot_pos = tp + fp
    return tp / tot_pos if tot_pos else 0

print(f"Sim: {sim_bayes()*100:.1f}%")
```

```
import math

def entropy(p_h):
    if p_h == 0 or p_h == 1:
        return 0
    p_t = 1 - p_h
    return - (p_h * math.log2(p_h) + p_t * math.log2(p_t))

print(f"A: {entropy(0.5):.2f}")
print(f"B: {entropy(0.99):.2f}")
print(f"C: {entropy(0.01):.2f}")
```
