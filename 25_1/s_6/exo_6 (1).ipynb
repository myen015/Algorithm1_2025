**1. Class Definition for the Tree Object**

```python
class Node:
    def start(self, w):
        self.weight = w
        self.children = []

node1 = Node()
node1.start(1)
print("node weight:", node1.weight)
print("node children:", node1.children)
```

**2. Generate a tree of depth N = 3, with initial parent tree of weight 1/n**

**step 1 create function to make tree**

```python
def makeTree(depth, weight):
    if depth == 0:
        return None
    
    node = Node()
    node.start(weight)
    
    for i in range(3):
        child_weight = weight / 3
        child = makeTree(depth - 1, child_weight)
        if child != None:
            node.children.append(child)
    
    return node

print("function created")
```

**step 2 create tree with depth 3**

```python
root = makeTree(3, 1)
print("tree created")
print("root weight:", root.weight)
print("root has children:", len(root.children))
```

**3. create a depth first recursive function visiting each node and summing up the weights. Make sure it returns 1 for various n's!**

**step 1 create DFS function**

```python
def sumWeightsDFS(node):
    if node == None:
        return 0
    
    total = node.weight
    
    for i in range(len(node.children)):
        child = node.children[i]
        child_sum = sumWeightsDFS(child)
        total = total + child_sum
    
    return total

print("DFS function created")
```

**step 2 test DFS**

```python
result_dfs = sumWeightsDFS(root)
print("DFS total weight:", result_dfs)
```

**4. Same with breadth first, check also 1!**

**step 1 create BFS function**

```python
def sumWeightsBFS(node):
    if node == None:
        return 0
    
    total = 0
    queue = []
    queue.append(node)
    
    while len(queue) > 0:
        current = queue[0]
        queue.pop(0)
        
        total = total + current.weight
        
        for i in range(len(current.children)):
            child = current.children[i]
            queue.append(child)
    
    return total

print("BFS function created")
```

**step 2 test BFS**

```python
result_bfs = sumWeightsBFS(root)
print("BFS total weight:", result_bfs)
```

**5. Same as above for both searches, but each time reach a node, flip the value sign. Make sure you get 1 and -1 after both first and second searches (run a test with fixed n)!**

**step 1 DFS with flip function**

```python
def sumDFSflip(node, should_flip):
    if node == None:
        return 0
    
    current_weight = node.weight
    if should_flip == True:
        current_weight = -current_weight
    
    total = current_weight
    
    for i in range(len(node.children)):
        child = node.children[i]
        if should_flip == True:
            child_sum = sumDFSflip(child, False)
        else:
            child_sum = sumDFSflip(child, True)
        total = total + child_sum
    
    return total

print("DFS flip function created")
```

**step 2 test DFS flip positive**

```python
root1 = makeTree(3, 1)
result1 = sumDFSflip(root1, False)
print("DFS flip positive start:", result1)
```

**step 3 test DFS flip negative**

```python
root2 = makeTree(3, 1)
result2 = sumDFSflip(root2, True)
print("DFS flip negative start:", result2)
```

**step 4 BFS with flip function**

```python
def sumBFSflip(node, start_flip):
    if node == None:
        return 0
    
    total = 0
    queue = []
    flip_queue = []
    
    queue.append(node)
    flip_queue.append(start_flip)
    
    while len(queue) > 0:
        current = queue[0]
        current_flip = flip_queue[0]
        queue.pop(0)
        flip_queue.pop(0)
        
        current_weight = current.weight
        if current_flip == True:
            current_weight = -current_weight
        
        total = total + current_weight
        
        for i in range(len(current.children)):
            child = current.children[i]
            queue.append(child)
            if current_flip == True:
                flip_queue.append(False)
            else:
                flip_queue.append(True)
    
    return total

print("BFS flip function created")
```

**step 5 test BFS flip positive**

```python
root3 = makeTree(3, 1)
result3 = sumBFSflip(root3, False)
print("BFS flip positive start:", result3)
```

**step 6 test BFS flip negative**

```python
root4 = makeTree(3, 1)
result4 = sumBFSflip(root4, True)
print("BFS flip negative start:", result4)
```

**6. Write recursive and non recursive versions of breadth first**

**step 1 create recursive BFS**

```python
def sumBFSrecursive(queue, total):
    if len(queue) == 0:
        return total
    
    current = queue[0]
    queue.pop(0)
    
    total = total + current.weight
    
    for i in range(len(current.children)):
        child = current.children[i]
        queue.append(child)
    
    return sumBFSrecursive(queue, total)

print("recursive BFS function created")
```

**step 2 test recursive BFS**

```python
root5 = makeTree(3, 1)
queue_test = []
queue_test.append(root5)
result5 = sumBFSrecursive(queue_test, 0)
print("BFS recursive total:", result5)
```

**7. Whilst depth first can be implemented recursively, it is not recommended for breadth first! Explain why?**

Answer:
DFS is good for recursion because it uses a stack like function calls, so it goes deep first. This means it uses less memory when the tree is not very wide. It keeps going down into one path until it hits the end, so it doesn't need to store all the nodes on each level.
But BFS is not good for recursion because it uses a queue, which needs more memory when the tree is wide. At each level, there can be many nodes, and since BFS works by checking all the nodes at the current level before moving to the next one, you have to store all the nodes in the queue, which takes a lot of memory.

Example:
Lets say you have a tree with 10 levels, and each level has 100 nodes. If you are doing BFS recursively, at each level, you need to store all 100 nodes in memory. So at level 1, you store 100 nodes, at level 2, another 100 nodes, and so on. This can quickly use up a lot of memory.
For DFS you don’t store all the nodes at once. You only store the nodes on the current path you’re exploring, and when you go deeper into the tree, you discard the nodes from the previous level. So it uses less memory, especially for deep but narrow trees
