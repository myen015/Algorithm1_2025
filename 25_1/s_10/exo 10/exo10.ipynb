```python
import math

# Problem 2
def bayes():
    diseaseP = 0.001
    sensitivity = 0.99
    specificity = 0.99

    false_pos = 1-specificity

    # Bayes
    numerator = sensitivity*diseaseP
    denominator = numerator+false_pos*(1-diseaseP)
    diseaseIfPos = numerator/denominator

    print(f"Promlem2. Probability = {diseaseIfPos} ({diseaseIfPos*100:.1f}%)")
    print()

bayes()
```

```python
# Problem 3
def shannon(p):
    q = 1-p
    if p==0 or q==0: return 0
    return -(p*math.log2(p) + q*math.log2(q))

def entropy():
    coins = {"A": 0.5, "B": 0.99, "C": 0.01}

    print("Problem3. Shannon Entropy:")
    for name, p in coins.items():
        entropy = shannon(p)
        print(f"Coin {name}: p(heads) = {p:.2f}, H = {entropy:.3f} bits")
    print()


entropy()
print("P - polynomial time solutions\nNP-complete - polynomial verification, no eff.solution\n9% - rare disease, many false positives\nMore predictable system = fewer bits of entropy")
```
