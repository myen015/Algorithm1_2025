```python
class TreeNode:
    def __init__(self, weight, n):
        self.weight = weight
        self.n = n
        self.children = []

    def generate_children(self):
        child_weight = self.weight / self.n
        for _ in range(self.n):
            self.children.append(TreeNode(child_weight, self.n))

# has a weight
# has n children
# each child has weight = parent.weight / n
```

```python
def build_tree(node, depth):
    if depth == 0:
        return
    node.generate_children()
    for child in node.children:
        build_tree(child, depth - 1)

# Example
n = 4
root = TreeNode(1, n)
build_tree(root, depth=3)

# Initial parent weight = 1
# Each child weight = 1 / n
```

```python
#sum of weights

def dfs_sum(node):
    total = node.weight
    for child in node.children:
        total += dfs_sum(child)
    return total
```

```python
from collections import deque

def bfs_sum(root):
    queue = deque([root])
    total = 0

    while queue:
        node = queue.popleft()
        total += node.weight
        for child in node.children:
            queue.append(child)

    return total
```

```python
#  BFS should NOT be recursive

def bfs_iterative(root):
    queue = deque([root])
    result = []

    while queue:
        node = queue.popleft()
        result.append(node.weight)
        for child in node.children:
            queue.append(child)

    return result

# but i wrote recursive vertion too

def bfs_recursive(queue, result):
    if not queue:
        return
    node = queue.popleft()
    result.append(node.weight)
    for child in node.children:
        queue.append(child)
    bfs_recursive(queue, result)

# Usage
result = []
bfs_recursive(deque([root]), result)
```
