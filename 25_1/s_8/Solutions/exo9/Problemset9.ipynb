Problem 2. Equivalence of NAND to AND, OR, and NOT (4/10 pts)

The NAND operation produces an output of 0 only when both inputs are 1, and outputs 1 otherwise.

From this single operation, it is possible to construct all basic Boolean operations:

NOT can be obtained by applying NAND to the same input twice.

AND can be obtained by first applying NAND to the inputs and then negating the result using another NAND gate.

OR can be obtained by first negating both inputs using NAND gates and then applying NAND to the results.

Since AND, OR, and NOT form a complete set of Boolean operations, and all of them can be built using only NAND gates, it follows that NAND is a universal operation.
