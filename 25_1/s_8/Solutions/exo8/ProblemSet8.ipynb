Fundamental Algorithm Techniques – Problem Set #8
Problem 1. Strongly Connected Components and Reversal (4/10 pts)
1. Algorithm to compute the reversal of a directed graph in O(V + E) time

To compute the reversal of a directed graph, we construct a new graph with the same set of vertices. For every directed edge from a vertex u to a vertex v in the original graph, we add a directed edge from v to u in the reversed graph.

This process requires scanning each vertex once and each edge exactly once. Since each edge is reversed in constant time, the total running time is linear in the number of vertices and edges, that is O(V + E).

2. Proof that the strong component graph scc(G) is acyclic

The strong component graph scc(G) is obtained by collapsing each strongly connected component of G into a single vertex, and keeping edges between components if there is at least one edge between the corresponding vertices in G.

Assume, for contradiction, that scc(G) contains a directed cycle. This would mean that there are several distinct strongly connected components that are reachable from each other in a cycle. However, this implies that every vertex in these components can reach every other vertex, which contradicts the definition of strongly connected components as maximal sets.

Therefore, scc(G) cannot contain any directed cycles and must be acyclic.

3. Proof that scc(rev(G)) = rev(scc(G))

Reversing a graph does not change which vertices can mutually reach each other. Therefore, the strongly connected components of G and rev(G) are identical as sets of vertices.

When the direction of all edges is reversed, the edges between the strongly connected components are also reversed. As a result, the strong component graph of the reversed graph is exactly the reversal of the strong component graph of the original graph.

Hence, scc(rev(G)) equals rev(scc(G)).

4. Reachability in G and in scc(G)

Let S(v) denote the strongly connected component containing vertex v.

If a vertex u can reach a vertex v in the original graph G, then there exists a path from u to v. This path may pass through several strongly connected components. When each strongly connected component is contracted into a single vertex, the path becomes a path from S(u) to S(v) in scc(G).

Conversely, if there is a path from S(u) to S(v) in scc(G), then by expanding each component, there exists a corresponding path from u to v in the original graph G.

Thus, u can reach v in G if and only if S(u) can reach S(v) in scc(G).

Problem 2. Euler Tour (2/10 pts)
1. Condition for the existence of an Euler tour

A strongly connected directed graph has an Euler tour if and only if, for every vertex, the in-degree is equal to the out-degree.

If an Euler tour exists, every time the tour enters a vertex, it must also leave it, which implies that the number of incoming edges equals the number of outgoing edges.

Conversely, if the graph is strongly connected and each vertex has equal in-degree and out-degree, it is always possible to construct a cycle that traverses every edge exactly once.

2. O(E)-time algorithm to find an Euler tour

An Euler tour can be found by first selecting any vertex and following unused outgoing edges until returning to the starting vertex, forming a cycle. If there are still unused edges, another cycle can be found starting from a vertex on the existing cycle that has unused edges.

These cycles are then merged into a single tour. Since each edge is used exactly once and each edge is processed a constant number of times, the total running time of this algorithm is O(E).

Problem 3. Topological Search (4/10 pts)

We are given the directed graph with edges:
A → B, A → C, B → C, B → D, C → E, D → E, D → F, G → F, G → E.

Topological sort starting from A

Vertex A must appear before B and C.
Vertex B must appear before C and D.
Vertex C must appear before E.
Vertex D must appear before E and F.
Vertex G must appear before E and F.

One valid topological ordering is:
A, B, C, D, G, E, F

Topological sort starting from another node

Starting from G, another valid ordering is:
G, A, B, D, C, F, E

Conclusion

There are multiple valid topological orderings for this graph because several vertices are independent of each other. Topological sorting produces one of many possible linear orders that respect all precedence constraints.
