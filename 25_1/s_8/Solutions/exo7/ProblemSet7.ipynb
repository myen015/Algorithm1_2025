Problem 1. Graph Play
1. Directed graphs and their transposed graphs

A directed graph consists of vertices connected by directed edges.
The transpose of a directed graph is obtained by reversing the direction of every edge while keeping the same set of vertices.

For example, if the original graph contains an edge from vertex A to vertex B, then the transposed graph will contain an edge from B to A. No other changes are made to the graph structure.

2. Undirected graphs and their inverse (complement) graphs

For undirected graphs, the inverse graph usually refers to the complement graph.

The complement of an undirected graph is constructed by connecting two distinct vertices if and only if there is no edge between them in the original graph. Self-loops are not allowed.

As a result, all missing edges in the original graph appear in the inverse graph, and all original edges are removed.

3. What happens if the original graph is dense

If the original graph is dense, meaning it contains almost all possible edges, then its inverse (complement) graph will be sparse and contain very few edges.

Conversely, if the original graph is sparse, its inverse graph will be dense. Thus, graph density is inverted when taking the complement.

4. Very simple undirected graphs and their dual graphs

The dual graph is defined for planar graphs, which are graphs that can be drawn on a plane without edge crossings.

In a dual graph, each face of the planar graph is represented as a vertex. Two vertices in the dual graph are connected by an edge if the corresponding faces in the original graph share a common edge.

For example, a simple triangle has an inner face and an outer face. These two faces become vertices in the dual graph, connected by an edge.

5. Why the dual graph is only well-defined for planar graphs

The definition of a dual graph relies on the existence of well-defined faces, which only occur when a graph has a planar embedding.
