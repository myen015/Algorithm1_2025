## Problem 1: Classify Problems by Complexity Class

Classify each problem as P, NP, NP-complete, or NP-hard.

```python
# Problem 1: Complexity Classification

problems = {
    "1. find max, linear search, shortest path in unweighted graph, matrix multiplication": "P",
    "2. sorting of list, Dijkstra on non-negative weights, BFS, DFS, merge sort, quicksort": "P",
    "3. sudoku": "NP-complete",
    "4. 3 coloring of graph, scheduling with conflicts": "NP-complete",
    "5. Traveling Salesperson Problem, Hamiltonian Cycle, Clique": "NP-complete",
    "6. Cryptography, factoring large integers": "NP",
    "7. Halting Problem, busy beaver": "Undecidable (beyond NP-hard)"
}

print("Classification of Computational Problems:\n")
print("=" * 80)
for problem, complexity in problems.items():
    print(f"{problem}")
    print(f"  → {complexity}\n")

print("\n" + "=" * 80)
print("\nSummary:")
print("P: Problems solvable in polynomial time")
print("NP: Problems verifiable in polynomial time")
print("NP-complete: Hardest problems in NP (if one is in P, all NP problems are)")
print("NP-hard: At least as hard as NP-complete (may not be in NP)")
print("Undecidable: No algorithm can solve all instances")
```

## Problem 2: Bayes' Theorem - Medical Testing

A deadly disease affects 0.1% of people. A test is 99% accurate (both sensitivity and specificity). A patient tests positive. What is the probability they actually have the disease?

```python
# Problem 2: Bayes' Theorem Application

# Given information
P_disease = 0.001  # 0.1% of people have the disease
P_no_disease = 1 - P_disease  # 99.9% don't have the disease

sensitivity = 0.99  # P(positive | disease) = 99%
specificity = 0.99  # P(negative | no disease) = 99%

P_positive_given_disease = sensitivity
P_positive_given_no_disease = 1 - specificity  # False positive rate = 1%

# Calculate P(positive) using law of total probability
P_positive = (P_positive_given_disease * P_disease +
              P_positive_given_no_disease * P_no_disease)

# Apply Bayes' Theorem: P(disease | positive)
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print("Problem 2: Medical Test Probability\n")
print("=" * 60)
print(f"Disease prevalence: {P_disease * 100}%")
print(f"Test accuracy (sensitivity & specificity): {sensitivity * 100}%")
print(f"\nPatient tests POSITIVE")
print("=" * 60)
print(f"\nP(positive) = {P_positive:.6f} or {P_positive * 100:.4f}%")
print(f"\n✓ P(disease | positive) = {P_disease_given_positive:.6f}")
print(f"  = {P_disease_given_positive * 100:.2f}%")
print(f"\n≈ 9% (approximately 9.01%)")
print("\n" + "=" * 60)
print("\nExplanation:")
print("Even with a 99% accurate test, the low disease prevalence (0.1%)")
print("means most positive results are false positives.")
print(f"\nOut of 100,000 people:")
print(f"  - {int(P_disease * 100000)} have disease → {int(P_disease * 100000 * sensitivity)} test positive")
print(f"  - {int(P_no_disease * 100000)} don't have disease → {int(P_no_disease * 100000 * P_positive_given_no_disease)} test positive (false)")
print(f"  - Probability = {int(P_disease * 100000 * sensitivity)}/{int(P_disease * 100000 * sensitivity + P_no_disease * 100000 * P_positive_given_no_disease)} ≈ 9%")
```

## Problem 3: Shannon Entropy Analysis

Calculate Shannon entropy for three coins with different probabilities of heads.

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Problem 3: Shannon Entropy

def shannon_entropy(probabilities):
    """
    Calculate Shannon entropy: H(X) = -Σ p_i * log2(p_i)
    """
    entropy = 0
    for p in probabilities:
        if p > 0:  # Handle log(0) case
            entropy -= p * math.log2(p)
    return entropy

# Define the three coins
coins = {
    "A (Fair coin)": {"P(Heads)": 0.50, "Surprise": "Medium"},
    "B (Biased coin)": {"P(Heads)": 0.99, "Surprise": "Tiny"},
    "C (Heavily biased)": {"P(Heads)": 0.01, "Surprise": "HUGE!"}
}

print("Problem 3: Shannon Entropy Analysis\n")
print("=" * 70)
print("Formula: H(X) = -Σ p_i * log₂(p_i)\n")

results = {}

for coin_name, coin_data in coins.items():
    p_heads = coin_data["P(Heads)"]
    p_tails = 1 - p_heads

    # Calculate entropy
    entropy = shannon_entropy([p_heads, p_tails])
    results[coin_name] = entropy

    print(f"Coin {coin_name}:")
    print(f"  P(Heads) = {p_heads:.0%}")
    print(f"  P(Tails) = {p_tails:.0%}")
    print(f"  Surprise when heads: {coin_data['Surprise']}")
    print(f"  H(X) = {entropy:.6f} bits")
    print(f"  H(X) ≈ {entropy:.2f} bits\n")

print("=" * 70)
print("\nKey Insights:")
print(f"1. Fair coin (50/50): Entropy = {results['A (Fair coin)']:.2f} bits (maximum uncertainty)")
print(f"2. 99% biased coin: Entropy = {results['B (Biased coin)']:.4f} bits (very predictable)")
print(f"3. 1% biased coin: Entropy = {results['C (Heavily biased)']:.4f} bits (very predictable)")
print("\nWhy is a fair coin worth 1 bit?")
print("  → Maximum uncertainty = maximum information content")
print("  → Knowing the outcome resolves 1 full bit of uncertainty")
print("\nWhy is a 99% biased coin only 0.08 bits?")
print("  → Outcome is almost certain (very predictable)")
print("  → Learning the outcome provides very little new information")
print("  → The 'surprise' when heads doesn't appear is huge, but it's rare!")
```

```python
# Visualize entropy as a function of probability

p_values = np.linspace(0.001, 0.999, 1000)
entropy_values = [shannon_entropy([p, 1-p]) for p in p_values]

plt.figure(figsize=(12, 6))

# Plot entropy curve
plt.plot(p_values, entropy_values, 'b-', linewidth=2, label='Entropy H(X)')

# Mark the three coins
plt.plot(0.50, results['A (Fair coin)'], 'go', markersize=12, label='Coin A (50%): 1.00 bits')
plt.plot(0.99, results['B (Biased coin)'], 'ro', markersize=12, label='Coin B (99%): 0.08 bits')
plt.plot(0.01, results['C (Heavily biased)'], 'mo', markersize=12, label='Coin C (1%): 0.08 bits')

# Add vertical lines
plt.axvline(0.50, color='g', linestyle='--', alpha=0.3)
plt.axvline(0.99, color='r', linestyle='--', alpha=0.3)
plt.axvline(0.01, color='m', linestyle='--', alpha=0.3)

plt.xlabel('P(Heads)', fontsize=12)
plt.ylabel('Entropy (bits)', fontsize=12)
plt.title('Shannon Entropy vs. Coin Bias\n(Maximum entropy at P=0.5)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)
plt.xlim(0, 1)
plt.ylim(0, 1.1)

plt.tight_layout()
plt.show()
```
