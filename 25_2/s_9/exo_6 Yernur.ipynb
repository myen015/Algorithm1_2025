```python
from collections import deque

class Node:
    def __init__(self, w):
        self.w = w
        self.c = []

def build(depth, w, n):
    r = Node(w)
    if depth == 0:
        return r
    for _ in range(n):
        r.c.append(build(depth - 1, w / n, n))
    return r

def dfs(r):
    s = r.w
    for x in r.c:
        s += dfs(x)
    return s

def bfs(r):
    q = deque([r])
    s = 0
    while q:
        x = q.popleft()
        s += x.w
        for y in x.c:
            q.append(y)
    return s

def dfs_flip(r):
    r.w = -r.w
    s = r.w
    for x in r.c:
        s += dfs_flip(x)
    return s

def bfs_flip(r):
    q = deque([r])
    s = 0
    while q:
        x = q.popleft()
        x.w = -x.w
        s += x.w
        for y in x.c:
            q.append(y)
    return s

def bfs_recursive(level):
    if not level:
        return 0
    next_level = []
    s = 0
    for x in level:
        s += x.w
        for y in x.c:
            next_level.append(y)
    return s + bfs_recursive(next_level)


# Main
if __name__ == "__main__":
    n = 5
    depth = 3
    root_weight = 1.0 / (depth + 1)

    root = build(depth, root_weight, n)

    print("DFS sum:", dfs(root))
    print("BFS sum:", bfs(root))

    print("DFS flip:", dfs_flip(root))
    print("BFS flip:", bfs_flip(root))

    print("DFS flip again:", dfs_flip(root))
    print("BFS flip again:", bfs_flip(root))

    print("Recursive BFS sum:", bfs_recursive([root]))
```
